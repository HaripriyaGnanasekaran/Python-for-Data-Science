# DATA NORMALIZATIONS

We will normalize the data using three procedures.  

simple feature scaling.  

$x = x/\max(x)$  

min-max scaling.  

$x = \frac{x - \min(x)}{ \max(x) -\min(x)}$  

z-score scaling

$x = \frac{x - \mu}{\sigma}$

we will follow the min-max scaling as it will make sure we have values between 0-1

```python
cleandata['length']=(cleandata['length']-cleandata['length'].min())/(cleandata['length'].max()-cleandata['length'].min())
```

```python
cleandata['length']
```

```python
cleandata['width']=(cleandata['width']-cleandata['width'].min())/(cleandata['width'].max()-cleandata['width'].min())
```

```python
cleandata['height']=(cleandata['height']-cleandata['height'].min())/(cleandata['height'].max()-cleandata['height'].min())
```
